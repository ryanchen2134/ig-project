```python
import pandas as pd
```

```python
fd = pd.read_csv('influencer.csv')

#remove the first column
fd = fd.drop(fd.columns[0], axis=1)
#remove Name, Subscribers, Audience country, Authentic engagement\n, Engaement average\n
fd = fd.drop([' Name', 'Subscribers', 'Audience country', 'Authentic engagement\n', 'Engagement average\n'], axis=1)
fd
```

```python
#for every "NaN" on BOTH Category_1 and Category_2 of a user, drop them from the list
fd = fd.dropna(subset=['Category_1', 'Category_2'], how='all')
fd
```

```python
#for every science on either Category_1 and Category_2 of a user, drop them from the list
fd = fd[~fd['Category_1'].str.contains('Science', na=False)]
fd
```

```python
fd = fd.drop(['Category_1', 'Category_2'], axis=1)
fd
```

```python
fd.to_csv('influencers_only.csv', index=False)
```
