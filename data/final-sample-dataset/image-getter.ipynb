```python
# df[0:1]['link'].values[0] 

#download the link (its an image), and name it with the value of the first column: df[0:1]['shortcode'].values[0]  ."extension"

# df[0:1]['link'].values[0]
# df[0:1]['shortcode'].values[0]
# df[0:1]['link'].values[0].split('.')[-1]

def download_image(idx,url, name):
    import requests
    from pathlib import Path

    # Create a directory to save the images if it doesn't exist
    # Path('images').mkdir(parents=True, exist_ok=True)
    if not Path('images').exists():
        Path('images').mkdir(parents=True, exist_ok=True)

    # If the name already exists, skip downloading
    if Path(f'images/{name}').exists():
        print(f"({idx}) {name} already exists, skipping download.")
        return

    # Download the image
    response = requests.get(url)
    if response.status_code == 200:
        with open(f'images/{name}', 'wb') as f:
            f.write(response.content)
        print(f"({idx}) Downloaded {name}")
    else:
        print(f"({idx}) Failed to download {url}")
        
        
# for i in range(len(df)):
for idx, i in enumerate(df.index):
    # Get the URL and name for each image
    url = df[i:i+1]['head_image_url'].values[0]
    name = f"{df[i:i+1]['shortcode'].values[0]}.jpg"
    
    # Download the image
    download_image(idx,url, name)
```
